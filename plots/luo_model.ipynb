```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/luo_model/exp1/output_U.csv')

time = df.iloc[:, 0]

u_values = df.iloc[:, 1:].values

n = 50
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.heatmap(
    u_values.T,
    cmap='hot',
    xticklabels=200,
    yticklabels=np.round(np.linspace(0, 1, n + 1), 2),
    cbar_kws={'label': 'Density'},
   
)

# èŽ·å–åæ ‡è½´
ax = plt.gca()

ax.invert_yaxis()


u0_values = u_values[:, 0]
ax.plot(
    np.arange(len(time)), 
    u0_values * n,  
    color='cyan',
    linewidth=2,
    label='u0 (all type I group)'
)


u50_values = u_values[:, n]
ax.plot(
    np.arange(len(time)), 
    u50_values * n,
    color='lime',
    linewidth=2,
    label='u50 (all type G group)'
)


ax.legend(
    loc='upper left',
    bbox_to_anchor=(1.2, 1), 
    borderaxespad=0.,
    frameon=True  
)

fig.subplots_adjust(right=0.8)
plt.xlabel('Time step')
plt.ylabel('Fraction of type G individuals in group')
plt.title('Dynamics of group composition over time')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('../data/luo_model/exp2/output_U.csv')

time = df.iloc[:, 0]

u_values = df.iloc[:, 1:].values

n = 50

fig, ax = plt.subplots(figsize=(10, 7))
vmin = np.nanmin(u_values)
vmax = np.nanmax(u_values)

sns.heatmap(
    u_values.T,
    cmap='hot',
    xticklabels=200,
    yticklabels=np.round(np.linspace(0, 1, n + 1), 2),
    cbar_kws={'label': 'Density'},
    vmin=vmin,
    vmax=vmax
)

# èŽ·å–åæ ‡è½´
ax = plt.gca()

ax.invert_yaxis()


u0_values = u_values[:, 0]
ax.plot(
    np.arange(len(time)), 
    u0_values * n,  
    color='cyan',
    linewidth=2,
    label='u0 (all type I group)'
)


u50_values = u_values[:, n]
ax.plot(
    np.arange(len(time)), 
    u50_values * n,
    color='lime',
    linewidth=2,
    label='u50 (all type G group)'
)


ax.legend(
    loc='upper left',
    bbox_to_anchor=(1.2, 1), 
    borderaxespad=0.,
    frameon=True  
)

fig.subplots_adjust(right=0.8)
plt.xlabel('Time step')
plt.ylabel('Fraction of type G individuals in group')
plt.title('Dynamics of group composition over time')

plt.tight_layout()
plt.show()
```
